## Importing important libraries

```python
import cv2 
import matplotlib.pyplot as plt 
import streamlit as st
```

```python
image=cv2.imread('pill.png',0)
plt.imshow(image,cmap='gray')
```

```python
blur=cv2.GaussianBlur(image,(5,5),1)
plt.imshow(blur,cmap='gray')
```

```python
edge=cv2.Canny(blur,50,150)
plt.imshow(edge,cmap='gray')
```

```python
dilated=cv2.dilate(edge,None,iterations=2)
plt.imshow(dilated,cmap='gray')
```

```python

cnt,heirachy=cv2.findContours(dilated,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
print('Number of pills in the image is:',len(cnt))
```
